# Quick EDA — battery time-series surrogate dataset

Lightweight exploration only. The notebook is **not** part of the training pipeline;
use the scripts in `scripts/` for anything reproducible.

Run this notebook from the project root so `import src...` resolves.

```python
import sys
from pathlib import Path

# Ensure the project root (parent of notebooks/) is importable.
ROOT = Path.cwd()
if (ROOT / 'src').is_dir():
    sys.path.insert(0, str(ROOT))
elif (ROOT.parent / 'src').is_dir():
    sys.path.insert(0, str(ROOT.parent))

import numpy as np
import matplotlib.pyplot as plt

from src.data import discover_cases, load_aligned_case_data

DATA_ROOT = 'generate_training_data'
cases = discover_cases(DATA_ROOT)
cases
```

```python
case = load_aligned_case_data(DATA_ROOT, cases[0])
print('case_id      :', case.case_id)
print('n_samples    :', case.n_samples)
print('n_parameters :', case.n_parameters)
print('t_last       :', case.t_last)
print('V range      :', case.V.min(), case.V.max())
print('T range      :', case.T.min(), case.T.max())
```

```python
# Plot a handful of voltage and temperature curves.
rng = np.random.default_rng(0)
rows = rng.choice(case.n_samples, size=min(8, case.n_samples), replace=False)

fig, axes = plt.subplots(1, 2, figsize=(13, 4))
for r in rows:
    axes[0].plot(case.time_s, case.V[r], alpha=0.7)
    axes[1].plot(case.time_s, case.T[r], alpha=0.7)
axes[0].set(title='Voltage', xlabel='time [s]', ylabel='V')
axes[1].set(title='Temperature', xlabel='time [s]', ylabel='C')
for ax in axes:
    ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
